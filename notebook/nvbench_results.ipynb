```python
from nebula import root
from nebula.common import read_pickle
import os.path as osp
import matplotlib.pyplot as plt
```

```python
def plot(res, title):
    plt.plot([int(i+1) for i in range(res["epoch"])], res["train_loss"], label = "Train Loss")
    plt.plot([int(i+1) for i in range(res["epoch"])], res["valid_loss"], label = "Validation Loss")
    plt.title(title)
    plt.legend()
    plt.show()
```

Comments:
    
    3 models are trained:
    
    1. original nc net model (multihead transformer sequence to sequence model, with attention forcing technique developed by the original team)
    2. nc net model without attention forcing
    3. new sequence to sequence model using bert as encoder.
    
    the nc net model with and without attention forcing show similar results, suggesting small improvement due to attention forcing
    
    the bert model shows a higher loss, but most likely due to its generalizability (the original ncnet model contains only vocab that exist in the data set)
    
    all three model overfits after the first epoch, suggesting that results after first epoch may not be relevant.
    
    comparing only first epoch:
    
    1. ncnet,       train loss: 1.78, valid loss: 0.56
    2. ncnet no af, train loss: 1.79, valid loss: 0.56
    3. nv bert,     train loss: 1.02, valid loss: 0.83

```python
# train results of ncnet model
# (sequence to sequence multihead transformer model with attention forcing)
res_ncnet_af = read_pickle(osp.join(root(), "model", "ncnet", "result", "train_results.pkl"))
```

```python
plot(res_ncnet_af, "ncNet train results")
```
